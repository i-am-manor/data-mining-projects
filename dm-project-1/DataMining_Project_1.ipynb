```python
df["Date"] = pd.to_datetime(df["Date"], format = '%d/%m/%Y')

df["Date"][0]
```

```python
# extracting the Year out of the date
df['Year'] = df['Date'].dt.year

df.head()
```

```python
# extracting the Month out of the date
df['Month'] = df['Date'].dt.month

df.head()
```

```python
# extracting the Weekday out of the date
df['Weekday'] = df['Date'].dt.weekday

df.head()
```

```python
# Delete "Date" column
df = df.drop('Date', axis=1)

df.head()
```

2.1. توجه داشته باشید که فقط مقادیر 1 تا 21 کدهای موجود در جدول ICD را نشان میدهد.از آنجایی که تمام مقادیر موجود در ICD نشاندهنده یک نوع بیماری است،
یک متغیر binary جدید تحت عنوان Disease تعریف کنید که بررسی کند دلیل غیبت  برای هر شخص در بازه ICD قرار دارد یا خیر.

```python
def in_icd(val):
  if 1 <= val <= 21:
    return 1
  else:
    return 0

df["Disease"] = df["Absence_Reason"].apply(in_icd)
df.head()
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(12, 6))

sns.violinplot(
    y="Absence_Duration_Hours",
    data=df[df['Social_Drinker_Status'] == 'Yes'],
    ax=ax1,
    color='red'
)
ax1.set_xlabel('Drinkers')
ax1.set_ylabel('Absence Duration (Hours)')
ax1.set_title('Distribution of Absence Hours for Drinkers')

sns.violinplot(
    y="Absence_Duration_Hours",
    data=df,
    ax=ax2,
)
ax2.set_xlabel('Employees')
ax2.set_ylabel('Absence Duration (Hours)')
ax2.set_title('Distribution of Absence Hours for All Employees')

sns.violinplot(
    y="Absence_Duration_Hours",
    data=df[df['Social_Smoker_Status'] == 'Yes'],
    ax=ax3,
    color='gray'
)
ax3.set_xlabel('Smokers')
ax3.set_ylabel('Absence Duration (Hours)')
ax3.set_title('Distribution of Absence Hours for Smokers')


plt.tight_layout()
plt.show()
```

<b> 3. فاکتورهای هزینه حمل و نقل و فاصله تا محل کار </b>
<p align="justify">
<font size=5>
دو شاخص ممکن برای غیبت ممکن است فاصله بین خانه و محل کار و هزینه‌های حمل و نقل (ستون هزینه حمل و نقل) باشد. کارمندانی که باید مسافت بیشتری را طی کنند یا هزینه‌های حمل و نقل به محل کار آن‌ها بالاتر است، ممکن است به غیبت بیشتری دچار شوند. در این بخش، رابطه بین این متغیرها و زمان غیبت به ساعت را بررسی خواهیم کرد.
<br>
 از آنجا که ما اعتقاد نداریم که عوامل مذکور ممکن است نشان‌دهنده مشکلات بیماری باشند به معنای دیگر به طور ذاتی عامل بیماری نباشد، ما رابطه ممکن با ستون دلیل غیبت را در نظر نخواهیم گرفت و به بررسی رابطه آن با زمان غیبت بسنده میکنیم.
<br>
1.3. با استفاده از joint plot رابطه بین فاصله محل اقامت تا کار و ساعات غیبت را نمایش دهید.
<br>
تحلیل بصری مشابهی برای ویژگی هزینه های حمل و نقل انجام دهید.
<br> درمورد نمودار jointplot و موارد کاربرد آن توضیح دهید.
</p>
</font>
</div>
</font>

```python
sns.jointplot(x="Distance_to_Work", y="Absence_Duration_Hours", kind = "reg", data=df)

plt.xlabel('Distance to Work')
plt.ylabel('Absence Duration (Hours)')

plt.show()
```

```python
sns.jointplot(x="Transportation_Expense", y="Absence_Duration_Hours", kind = "reg", data=df)

plt.xlabel('Transportation Expense')
plt.ylabel('Absence Duration (Hours)')

plt.show()
```

<b> <font color='red'> توضیحات پاسخ: </font></b>  نمودار jointplot، رابطه بین دو متغیر را نشان می‌دهد.
این نمودار از دو قسمت Main Plot و Marginal Plot تشکیل می‌شود.
<br>
Main Plot، رابطه میان دو متغیر که روی محور x و y قرار گرفته‌اند را نشان می‌دهد. از این قسمت می‌توان برای
کشف ماهیت رابطه بین دو متغیر کمک گرفت(این رابطه می‌تواند خطی، غیرخطی یا خوشه‌ای باشد).
<br>
قسمت Marginal Plot که روبه‌روی  هر یک از محورهای x و y قرار گرفته است؛
نحوه توزیع متغیری که روی این محور قرار دارد را نشان می‌دهد.
همچنین از روی این قسمت میتوان چولگی، کشیدگی و سایر ویژگی‌های توزیعی را استنباط کرد.

همانطور که مشاهده می‌شود، توزیع فواصل از محل اقامت تا محل کار و هزینه‌های حمل و نقل به نظر نزدیک به توزیع‌های نرمال می‌آید، در حالی که زمان غیبت به ساعت‌ها به شدت به سمت راست منحرف شده است. این موضوع باعث می‌شود مقایسه بین متغیرها دشوار باشد. یکی از راه‌حل‌های این مشکل، تبدیل داده‌ها به چیزی نزدیک به توزیع نرمال است. روشی موثر برای انجام این تبدیل، استفاده از تبدیل‌های Box-Cox یا Yeo-Johnson است که هر دو در کتابخانه scipy موجود است.

2.3. با توجه به اینکه ستون زمان غیبت شامل مقادیر صفر است از تبدیل Yeo-Johnson استفاده کنید.
<br>
سپس هر دو نمودار بالا را مجددا رسم کنید و با تحلیل این نمودارها درمورد تاثیرگذاری این دو فاکتور بر روی زمان غیبت توضیح دهید.
</p>

</div>
</font>

```python
Distance_to_Work_transformed, _ = stats.yeojohnson(df['Distance_to_Work'])
Absence_Duration_transformed, _ = stats.yeojohnson(df['Absence_Duration_Hours'])

sns.jointplot(x=Distance_to_Work_transformed, y=Absence_Duration_transformed, kind="reg", data=df)

plt.xlabel('Distance to Work (Yeo-Johnson transformed)')
plt.ylabel('Absence Duration (Yeo-Johnson transformed)')

plt.show()
```

```python
Transportation_Expense_transformed, _ = stats.yeojohnson(df['Transportation_Expense'])
Absence_Duration_transformed, _ = stats.yeojohnson(df['Absence_Duration_Hours'])

sns.jointplot(x=Transportation_Expense_transformed, y=Absence_Duration_transformed, kind="reg", data=df)

plt.xlabel('Transportation Expense (Yeo-Johnson transformed)')
plt.ylabel('Absence Duration (Yeo-Johnson transformed)')

plt.show()
```
